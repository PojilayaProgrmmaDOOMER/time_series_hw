```python
import matplotlib.pyplot as plt
import numpy as np
import random

myseed = 42
random.seed(myseed)
np.random.seed(myseed)

def relu(x):
    return np.maximum(0, x)

time = np.linspace(-5, 5, num=100)
N = 7
timeseries = {'time': time}
rand = np.random.random(size=7)

for i in range(N):
    timeseries[f'sin{i}'] = np.sin(time + rand[i])
for i in range(N):
    timeseries[f'cos{i}'] = np.cos(time + rand[i])
for i in range(N):
    timeseries[f'relu{i}'] = relu(time + rand[i])

for i in range(N):
    plt.plot(time, timeseries[f'sin{i}'], '-b')
for i in range(N):
    plt.plot(time, timeseries[f'cos{i}'], '-g')
for i in range(N):
    plt.plot(time, timeseries[f'relu{i}'], '-r')

plt.show()
```

```python
from scipy.spatial.distance import euclidean

euclidean_dist_custom = np.array([])

for i in range(1, 4):
    x = np.abs(timeseries[f"sin{i}"])
    y = np.abs(timeseries[f"sin{i - 1}"])
    distance = euclidean(x, y)
    euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
    print(f" euclidean   sin {i}  {i - 1}:", distance)

for i in range(1, 4):
    x = np.abs(timeseries[f"cos{i}"])
    y = np.abs(timeseries[f"cos{i - 1}"])
    distance = euclidean(x, y)
    euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
    print(f" euclidean   cos {i}  {i - 1}:", distance)

for i in range(1, 4):
    x = np.abs(timeseries[f"relu{i}"])
    y = np.abs(timeseries[f"relu{i - 1}"])
    distance = euclidean(x, y)
    euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
    print(f" euclidean   relu {i}  {i - 1}:", distance)

x = np.abs(timeseries["sin1"])
y = np.abs(timeseries["cos1"])
distance = euclidean(x, y)
euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
print(f" euclidean   sin1  cos1:", distance)

x = np.abs(timeseries["sin1"])
y = np.abs(timeseries["relu1"])
distance = euclidean(x, y)
euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
print(f" euclidean   sin1  relu1:", distance)

x = np.abs(timeseries["relu1"])
y = np.abs(timeseries["cos1"])
distance = euclidean(x, y)
euclidean_dist_custom = np.append(euclidean_dist_custom, distance)
print(f" euclidean   relu1  cos1:", distance)
```

```python
from dtaidistance import dtw

dtw_dist_custom = np.array([])

for i in range(1, 4):
    x = np.abs(timeseries[f"sin{i}"])
    y = np.abs(timeseries[f"sin{i - 1}"])
    distance = dtw.distance(x, y)
    dtw_dist_custom = np.append(dtw_dist_custom, distance)
    print(f" dtw   sin {i}  {i - 1}:", distance)

for i in range(1, 4):
    x = np.abs(timeseries[f"cos{i}"])
    y = np.abs(timeseries[f"cos{i - 1}"])
    distance = dtw.distance(x, y)
    dtw_dist_custom = np.append(dtw_dist_custom, distance)
    print(f" dtw   cos {i}  {i - 1}:", distance)

for i in range(1, 4):
    x = np.abs(timeseries[f"relu{i}"])
    y = np.abs(timeseries[f"relu{i - 1}"])
    distance = dtw.distance(x, y)
    dtw_dist_custom = np.append(dtw_dist_custom, distance)
    print(f" dtw   relu {i}  {i - 1}:", distance)

x = np.abs(timeseries["sin1"])
y = np.abs(timeseries["cos1"])
distance = dtw.distance(x, y)
dtw_dist_custom = np.append(dtw_dist_custom, distance)
print(f" dtw   sin1  cos1:", distance)

x = np.abs(timeseries["sin1"])
y = np.abs(timeseries["relu1"])
distance = dtw.distance(x, y)
dtw_dist_custom = np.append(dtw_dist_custom, distance)
print(f" dtw   sin1  relu1:", distance)

x = np.abs(timeseries["relu1"])
y = np.abs(timeseries["cos1"])
distance = dtw.distance(x, y)
dtw_dist_custom = np.append(dtw_dist_custom, distance)
print(f" dtw   relu1  cos1:", distance)
```

```python
#     
X = []
Y = []
for v in timeseries:
    if v != 'time':
        X.append(timeseries[v])
        c = v[:3]
        if c == "sin":
            Y.append(1)
        elif c == "cos":
            Y.append(2)
        else:
            Y.append(3)

#      
arr = np.arange(len(Y))
np.random.shuffle(arr)

X_train = [X[i] for i in arr[:-6]]
X_test = [X[i] for i in arr[len(X_train):]]
Y_train = [Y[i] for i in arr[:-6]]
Y_test = [Y[i] for i in arr[len(X_train):]]

# 
from pyts.classification import KNeighborsClassifier

clf = KNeighborsClassifier(metric='dtw')
clf.fit(X_train, Y_train)

print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} :")
    print(f"     {clf.predict_proba([X_test[i]])}")
    print(f"    {Y_test[i]}")
```

```python
# K     
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier()
clf.fit(X_train, Y_train)

print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} :")
    print(f"     {clf.predict_proba([X_test[i]])}")
    print(f"    {Y_test[i]}")

# Time Series Forest
from pyts.classification import TimeSeriesForest

clf = TimeSeriesForest(random_state=43)
clf.fit(X_train, Y_train)

print(f"Accuracy {clf.score(X_test, Y_test)}")
for i in range(len(Y_test)):
    print(f"{i + 1} :")
    print(f"     {clf.predict_proba([X_test[i]])}")
    print(f"    {Y_test[i]}")
```
